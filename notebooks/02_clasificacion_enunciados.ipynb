# 02 — Clasificación de enunciados

Clasifica cada párrafo del corpus limpio en:
DIAGNÓSTICO, PROPUESTA, CIFRA, NARRATIVA, ATAQUE_POLÍTICO, PROMESA_VAGA, MARCO_IDEOLÓGICO


```python
import re
import pandas as pd
from pathlib import Path

PROCESSED_DIR = Path('../data/processed')
df = pd.read_csv(PROCESSED_DIR / 'corpus_limpio.csv')
```

```python
# Clasificador por reglas (baseline — mejorar con anotación manual)

INDICADORES = {
    'CIFRA': [r'\$[\d\.,]+ (billones|millones)', r'[\d\.,]+ (hectáreas|casos|municipios|%|toneladas|personas)'],
    'PROPUESTA': [r'^(Mi gobierno|Crearé|Construiremos|Implementaremos|Lanzaremos|Estableceré|Vamos a|Habrá)', r'(primer día|primeros noventa días|plan de choque)'],
    'ATAQUE_POLÍTICO': [r'(Petro|Pacto Histórico|Iván Cepeda|extrema izquierda|gobierno actual)', r'(destruyó|fracasó|cómplice|saqueo|traición)'],
    'MARCO_IDEOLÓGICO': [r'(Doctrina Social|bien común|dignidad humana|Extrema Coherencia|principios y valores)', r'(Patria es|la Patria nos|restaurar la República)'],
    'DIAGNÓSTICO': [r'(Colombia (enfrenta|padece|vive|asiste a)|hoy (el país|Colombia|la nación))', r'(se estima|se registraron|según|datos)'],
    'PROMESA_VAGA': [r'(haremos una transformación|avanzaremos hacia|trabajaremos por|será prioritario)', r'(en el futuro|a mediano plazo|cuando sea necesario)'],
}

def clasificar(parrafo):
    for tipo, patrones in INDICADORES.items():
        for patron in patrones:
            if re.search(patron, parrafo, re.IGNORECASE):
                return tipo
    return 'NARRATIVA'

df['tipo_enunciado'] = df['parrafo'].apply(clasificar)
print(df['tipo_enunciado'].value_counts())
df.to_csv(PROCESSED_DIR / 'enunciados_clasificados.csv', index=False, encoding='utf-8')
```
